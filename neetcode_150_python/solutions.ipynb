```python
# https://leetcode.com/problems/contains-duplicate/description/
#217
#Type: EASY

#-=========Sol: 1================#
# Time complexity: O(n) -> for Loop
# Space complexity: O(n) -> hashset
class Solution(object):
    def containsDuplicate(self, nums):
        """
        :type nums: List[int]
        :rtype: bool
        """

        hashset = set()
        for num in nums:
            if num in hashset:
                return True
            hashset.add(num)
        return False

#==========Sol: 2================#
# Time complexity: O(n) -> O(n): set(nums), O(1): len(nums), O(1): len(set(nums))
# Space complexity: O(n) -> set(nums)
class Solution(object):
    def containsDuplicate(self, nums):
        return len(nums) != len(set(nums))
```
